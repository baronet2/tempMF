```python
import pandas as pd
from plotnine import *
```

```python
results = pd.DataFrame(
    {
        'method': ['PMF', 'PMF', 'TempMF', 'TempMF'],
        'dataset': ['train', 'test'] * 2,
        'rmse': [0.10197956948526844, 0.18090768218295747, 0.09215719007745264, 0.18038204567215332]
    }
)

results
```

```python
p = (
    results
    .assign(dataset = lambda d: pd.Categorical(d.dataset.str.capitalize(), ["Train", "Test"]))
    .pipe(ggplot, aes(x = 'method', y = 'rmse', fill = 'method')) +
    geom_bar(stat = 'identity') +
    facet_wrap("dataset") +
    theme_bw() +
    labs(y = "RMSE", x = "Method") +
    theme(figure_size=(10, 5), text=element_text(size=16), legend_entry_spacing_y=16, legend_position="none") +
    scale_fill_manual({'PMF': "#ddcab9", 'TempMF': '#6E5A4B'})
)

p.save("Results for Slides.png")

p
```
